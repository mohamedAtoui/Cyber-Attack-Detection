```python
# Function to round numerical values to a specified number of decimals and remove trailing zeros
def round_and_trim(column, decimals):
    return column.round(decimals).apply(lambda x: ('{0:.{1}f}'.format(x, decimals)).rstrip('0').rstrip('.'))

# Function to remove trailing zeros from numerical values
def remove_trailing_zeros(column):
    return column.apply(lambda x: ('{0:f}'.format(x)).rstrip('0').rstrip('.') if '.' in '{0:f}'.format(x) else '{0:f}'.format(x))

# Specify the columns that are categorical
category_col = ['originp', 'responp', 'fwd_data_pkts_tot', 'bwd_data_pkts_tot', 
                'fwd_header_size_tot', 'bwd_header_size_tot', 
                'fwd_subflow_pkts', 'bwd_subflow_pkts']

# Specify the columns that are numerical
num_col = ['flow_duration', 'fwd_pkts_tot', 'bwd_pkts_tot', 'fwd_pkts_per_sec', 
           'bwd_pkts_per_sec', 'flow_pkts_per_sec', 'down_up_ratio', 
           'fwd_pkts_payload.avg', 'fwd_pkts_payload.std', 'bwd_pkts_payload.avg', 
           'bwd_pkts_payload.std', 'fwd_iat.min', 'fwd_iat.max', 'bwd_iat.min', 
           'bwd_iat.max', 'payload_bytes_per_second', 'active.std']

# Convert the specified numerical columns to numeric data types
data1[num_col] = data1[num_col].apply(pd.to_numeric, errors='coerce')

# Convert the specified categorical columns to numeric data types
data1[category_col] = data1[category_col].apply(pd.to_numeric, errors='coerce')

# Apply rounding and trimming to numerical columns
for col in num_col:
    data1[col] = round_and_trim(data1[col], 3)

# Apply removal of trailing zeros to categorical columns
for col in category_col:
    data1[col] = remove_trailing_zeros(data1[col])
```

```python
y=data1.iloc[:,-1].values
len(y)
len(data1.iloc[:,-1])
np.unique(y)
```

```python
x=data1.iloc[:,:-1].values
x[0]
```

```python
# Import necessary libraries for model training, evaluation, and splitting the dataset
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Split the dataset into training and testing sets
# 80% of the data is used for training and 20% for testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=32)

# Initialize the K-Nearest Neighbors classifier with 31 neighbors
knn = KNeighborsClassifier(n_neighbors=31)

# Train the KNN classifier using the training data
knn.fit(x_train, y_train)

# Predict the labels for the test data
y_pred = knn.predict(x_test)

# Calculate the accuracy of the model by comparing the predicted labels with the actual labels
accuracy = accuracy_score(y_test, y_pred)

# Compute the confusion matrix to evaluate the accuracy of the classification
conf_matrix = confusion_matrix(y_test, y_pred)

# Generate a classification report to summarize the precision, recall, and F1-score of the classifier
class_report = classification_report(y_test, y_pred)

# Print the accuracy of the model
print(f'Accuracy: {accuracy}')

# Print the confusion matrix
print('Confusion Matrix:')
print(conf_matrix)

# Print the classification report
print('Classification Report:')
print(class_report)
```
